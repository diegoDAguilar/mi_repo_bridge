```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
pd.pivot_table(titanic, values='survived', index=['pclass'], columns=['sex'], aggfunc=np.sum)
```

```python
titanic.pivot_table(values='survived', index=['sex'], columns=['pclass'],
                    aggfunc={'survived':np.mean, 'fare':np.mean})
```

```python
titanic.groupby(['sex', 'pclass']).sum()
```

```python
titanic.groupby(['pclass', 'sex']).sum()
```
